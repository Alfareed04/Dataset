```python
 #file_name = "From GitHub.csv"

file_path = "Files/data/From GitHub.csv"
df = spark.read.csv(file_path, header=True, inferSchema=True)

df.show()
```

```python
from pyspark.sql.functions import col, when

df_transformed = df.withColumn("subject_1_status",when(col("subject_1") > 35, "PASS")
                 .otherwise("FAIL")
                 ).withColumn("subject_2_status",when(col("subject_2") > 35, "PASS")
                 .otherwise("FAIL")
                 ).withColumn("subject_3_status",when(col("subject_3") > 35, "PASS")
                 .otherwise("FAIL"))

df_transformed.show()
```

```python
from delta.tables import DeltaTable

df_transformed.write.option("overwriteSchema", "true").format("delta").mode("overwrite").save("Tables/student_data")
```
